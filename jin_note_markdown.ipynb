## Basic writing and formatting syntax
---

### Headings
---

* **Markdown code:**

```markdown
# Heading 1
## Heading 2
### Heading 3
#### Heading 4
##### Heading 5
###### Heading 6
```


* **Output:**

# Heading 1
## Heading 2
### Heading 3
#### Heading 4
##### Heading 5
###### Heading 6

### Styling Text
---

**Markdown code:**

```markdown
**Bold**
_Italic_
~~Strikethrough~~
**Bold _and_ italic**
```


**Output:**

**Bold**  
_Italic_  
~~Strikethrough~~  
**Bold _and_ italic**  

### Quoting Text
---

**Markdown code:**


```markdown
> Quote 1
>
>> Quote 2
>>
>>> Quote 3
>>
>>>> Quote 4
>>>
>>>>> Quote 5
```


**Output:**  
 
 
> Quote 1
>
>> Quote 2
>>
>>> Quote 3
>>
>>>> Quote 4
>>>
>>>>> Quote 5

### Quoting Code
---

**Markdown code:**


* Within a sentence

```markdown
Hello `Quote` world!
```


* Distinct block (triple backticks)

```markdown
Hello

` ` `
Quote
` ` `

World
```


**Output:**  


* Within a sentence

Hello `Quote` world!


* Distinct block (triple backticks)

Hello

```
Quote
```

World!

### Links
---


**Markdown code:**


```markdown
Link: [Jin's GitHub](https://github.com/ktj9279).
```


**Output:**  
 
 
Link: [Jin's GitHub](https://github.com/ktj9279).

### Section Links
---


* You can link directly to a section in a rendered file by hovering over the section heading to expose the link


※ 따로 코드를 추가할 필요는 없고, section heading 위에 마우스를 올렸을 때 우측에 뜨는 링크 아이콘을 클릭하면 해당 section의 시작 위치에 맞게 스크롤을 이동한다.

### Relative Links
---


* Repository 내에서 상대 경로(relative path)로 다른 파일의 링크를 걸 수 있다. 코드는 일반 링크 걸 때와 동일하고 소괄호 안에 상대 경로를 입력하면 된다. Repository를 clone하면 절대 경로(absolute path)로 지정한 링크에 문제가 생길 수 있기 때문에 relative link 사용을 추천.

### Lists
---

**Markdown code:**


* Unordered list

```markdown
* list 1
* list 2
* list 3
* list 4
```


or


```markdown
- list 1
- list 2
- list 3
- list 4
```


* Oredered list

```markdown
1. list 1
2. list 2
3. list 3
4. list 4
```


**Output:**  


* Unordered list


* list 1
* list 2
* list 3
* list 4


* Oredered list


1. list 1
2. list 2
3. list 3
4. list 4

## Organizing information with tables
---

### Creating a table
---

* A balnk line before a table
* Hyphens `-` to create each column's header
* Pipes `|` to separate each column


**Markdown code:**


```markdown

| Header 1    | Header 2    |
| ----------- | ----------- |
| Cell (1, 1) | Cell (1, 2) |
| Cell (2, 1) | Cell (2, 2) |
```


**Output:**  


| Header 1    | Header 2    |
| ----------- | ----------- |
| Cell (1, 1) | Cell (1, 2) |
| Cell (2, 1) | Cell (2, 2) |
